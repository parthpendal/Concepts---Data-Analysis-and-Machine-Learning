```python
try:
  source=requests.get("https://www.imdb.com/chart/top")
  # Raise an error if website is not reachable
  source.raise_for_status()

  soup=BeautifulSoup(source.text,'html.parser')
  movies=soup.find('tbody',class_='lister-list').find_all('tr')
  #print(movies)
  list1=list()
  for tr_tag in movies:
    rank=tr_tag.find('td',class_='titleColumn').get_text(strip=True).split(".")[0]
    Year=tr_tag.find('td',class_='titleColumn').span.text.split("(")[1].split(")")[0]
    name=tr_tag.find('td',class_='titleColumn').a.text
    rating=tr_tag.find('td',class_='ratingColumn imdbRating' ).strong.text
    #   .text
    #print(rank)
    #print(name)
    #print(Year)
    #print(rating)
    list1.append([rank,name,Year,rating])
except Exception as e:
  print(e)

df=pd.DataFrame(list1)
df.columns=['Rank','Movie Name','Year', 'Rating']
```
