```python
df1=df.drop(["customerID"],axis=1)
```

```python
df1.head()
```

```python
def find_cat_unique(df):
    for column_df in df.columns:
#   if df[column_df].dtypes == "object":
            print(column_df," : ",df[column_df].unique())

find_cat_unique(df1)
```

```python
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
find_cat_unique(df1)
```

```python
df1.replace('No',0,inplace=True)
df1.replace('Yes',1,inplace=True)
df1['gender'].replace('Male',1,inplace=True)
df1['gender'].replace('Female',0,inplace=True)
find_cat_unique(df1)
```

```python
df2=pd.get_dummies(data=df1,columns=['InternetService','Contract','PaymentMethod'])
```

```python
df2.head()
```

```python
find_cat_unique(df2)
```

```python
df2.dtypes
```

```python

df3= df2.loc[df2["TotalCharges"]!=' ']
df3["TotalCharges"]=df3["TotalCharges"].astype("float",errors='ignore')
```

```python
df3.dtypes
```

```python
from tensorflow import keras
import numpy as np
```

```python
df3.shape
```

```python
x_data=df3.drop(["Churn"],axis=1)
y_data=df3["Churn"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(x_data, y_data, test_size=0.2)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
def ANN(x_train,y_train,x_test,y_test,loss,weights):
    model = keras.Sequential ([
        keras.layers.Dense(10,input_shape=(26,),activation='relu'),
        keras.layers.Dense(10,activation='relu'),
        keras.layers.Dense(1,activation='sigmoid')
        ])
    model.compile(optimizer='adam',
             loss=loss,
             metrics=["accuracy"]
             )
    if weights== -1:
        model.fit(x_train,y_train,epochs=100)
    else:
        model.fit(x_train,y_train,epochs=100,class_weight=weights)
    model.evaluate(x_test,y_test)
    yp=model.predict(x_test)
    y_test_predict=[]
    for x in yp:
        if x>0.5:
            y_test_predict.append(1)
        else:
            y_test_predict.append(0)

    print(classification_report(y_test,y_test_predict))
```

```python
ANN(x_train,y_train,x_test,y_test,'binary_crossentropy',-1)
```

# Undersampling technique

```python

#Segregate class 0 and class 1
df3_class0=df3[df3['Churn']==0]
df3_class1=df3[df3['Churn']==1]
```

```python
# Get class0 count
df3_class0.Churn.value_counts()
```

```python
df3_class1.Churn.value_counts()
```

```python
# Understand how .sample works
df3_class0.sample(2)
```

```python
# Undersample class 0 to match with counts of class 1
df3_class0_under =df3_class0.sample(df3_class1.Churn.value_counts()[1])
```

```python
df3_class0_under.shape
```

```python
# Concatenate same count of  class 0 and class 1 records 
df_test_under=pd.concat([df3_class0_under,df3_class1],axis=0)
df_test_under.shape
```

```python
x_train,x_test, y_train,y_test=train_test_split(df_test_under.drop("Churn",axis=1),df_test_under["Churn"],test_size=0.2)
```

```python
ANN(x_train,y_train,x_test,y_test,'binary_crossentropy',-1)
```

# Oversampling technique

```python
df3_class0["Churn"].value_counts()[0]
```

```python
df3_class1["Churn"].value_counts()[1]
```

```python
df3_class1_over = df3_class1.sample(df3_class0["Churn"].value_counts()[0],replace=True)
df3_class1_over.shape
```

```python
# Concatenate same count of  class 0 and class 1 records 
df_test_under=pd.concat([df3_class0,df3_class1_over],axis=0)
df_test_under.shape
```

```python
x_train,x_test, y_train,y_test=train_test_split(df_test_under.drop("Churn",axis=1),df_test_under["Churn"],test_size=0.2)
ANN(x_train,y_train,x_test,y_test,'binary_crossentropy',-1)
```

# Using ENSAMBLE techinque

```python
x_train,x_test, y_train,y_test=train_test_split(df3.drop("Churn",axis=1),df3["Churn"],test_size=0.2)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
def ANN(x_train,y_train,x_test,y_test,loss,weights):
    model = keras.Sequential ([
        keras.layers.Dense(10,input_shape=(26,),activation='relu'),
        keras.layers.Dense(10,activation='relu'),
        keras.layers.Dense(1,activation='sigmoid')
        ])
    model.compile(optimizer='adam',
             loss=loss,
             metrics=["accuracy"]
             )
    if weights== -1:
        model.fit(x_train,y_train,epochs=100)
    else:
        model.fit(x_train,y_train,epochs=100,class_weight=weights)
    model.evaluate(x_test,y_test)
    yp=model.predict(x_test)
    y_test_predict=[]
    for x in yp:
        if x>0.5:
            y_test_predict.append(1)
        else:
            y_test_predict.append(0)

    print(classification_report(y_test,y_test_predict))
    return y_test_predict
```

```python
y_pred1=ANN(x_train[:1480],y_train[:1480],x_test,y_test,'binary_crossentropy',-1)
```

```python
y_pred2=ANN(x_train[1480:2960],y_train[1480:2960],x_test,y_test,'binary_crossentropy',-1)
```

```python
y_pred3=ANN(x_train[2961:4145],y_train[2961:4145],x_test,y_test,'binary_crossentropy',-1)
```
