# 1st method Fillna method


```python

df2=df1.fillna(0)
df2
```

```python
# Do fillna for specific columns
df3=df1.copy()
df3['temperature']=df1['temperature'].fillna(0)
df3['windspeed']=df1['windspeed'].fillna(0)
df3['event']=df1['event'].fillna('Normal')
```

```python
#Same thing can be achieved by different syntax
new_df=df1.fillna({'temperature':0,'windspeed':0,'event':'no event'})
```

# Second method forward fill & backward fill

```python

new_df1=df1.fillna(method='ffill')
new_df1
```

```python
#Backward fill
new_df2=df1.fillna(method='bfill')
new_df2
```

```python
#Limit keyword
df1
new_df2=df1.fillna(method='ffill',limit=1)
new_df2
```

# 3rd Method interpolate

```python
#linear interpoltion (average value)
print(df1)
new_df3=df1.interpolate()
new_df3
```

```python
#method=time
#linear interpoltion (average value)
print(df1)
new_df3=df1.interpolate(method="time")
new_df3
```

# 4th Method dropna

```python
#Dropping rows with na value
new_df4=df1.copy()
print(new_df4)
new_df4=new_df4.dropna(how="all")
new_df4
```

```python
#Dropping rows with all  value (except index)
print(df1)
new_df4=df1.dropna(how="all")
new_df4
```

```python
#Threshold paramneter
#Thereshold =1 means if if there is atleast 1 non na value keep the row
new_df4=df1.copy()
print(new_df4)
new_df4=new_df4.dropna(thresh=2)
new_df4
```
