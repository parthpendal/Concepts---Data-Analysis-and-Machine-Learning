```python
import cv2
import os
from google.colab.patches import cv2_imshow
img=cv2.imread("Persons4.jpg")
cv2_imshow(img)
```

Face detection is detects presence of a face in an image while face recognization involves identifying whose face it is.

Open CV comes with classifiers which helps in detection of face

Two important classifier are

1) Har cascades  https://github.com/opencv/opencv/tree/master/data/haarcascades (widely used but not that effective)

2)Local binary pattern (advanced classifier) - Delings face recognizer is used for most advance computer vision program

```python

#Convert image to gray
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)

```

```python
#Take xml from github and download in the local
#har_cascade=cv2.CascadeClassifier('C:\\Users\\parth\\anaconda3\\Lib\\site-packages\\cv2\\data\\haarcascade_frontalface_default.xml')
har_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
har_cascade
```

```python
faces_rec=har_cascade.detectMultiScale(gray,scaleFactor=1.1, minNeighbors=1)
print(faces_rec)
print(len(faces_rec))
```

```python
for (x,y,w,h) in faces_rec:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),thickness=2)

cv2_imshow(img)
```

We see 3 boxes rather than 2 boxes (there are 2 faces) due to noise

```python
faces_rec=har_cascade.detectMultiScale(gray,scaleFactor=1.1, minNeighbors=3)
print(faces_rec)
print(len(faces_rec))
for (x,y,w,h) in faces_rec:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),thickness=2)

cv2_imshow(img)
```
