```python
#for jupitor notebook
#cv2.imshow(first_read_image)

#for google colab
from google.colab.patches import cv2_imshow 
cv2_imshow(img)
```

```python
#Converting image to a grayscale
gray_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
#Blur an image - removing noise or bad light
blur=cv2.GaussianBlur(img,(1,1),cv2.BORDER_DEFAULT)
cv2_imshow(blur)
```

```python
#Edge cascade - trying to find edge in the image 
#img=cv2.imread("Boston_image.jpg")
cv2_imshow(img)
canny = cv2.Canny(img,125,127)
cv2_imshow(canny)
```

```python
#Dilating image
dilated=cv2.dilate(canny,(7,7),iterations=3)
cv2_imshow(dilated)
```

```python
#Eroding
eroded=cv2.erode(dilated,(7,7),iterations=3)
cv2_imshow(eroded)
```

```python
#resize image
resized=cv2.resize(img,(100,100))
cv2_imshow(resized)
```

```python
#Cropping image
cropped=resized[25:75,25:75]
cv2_imshow(cropped)
```
