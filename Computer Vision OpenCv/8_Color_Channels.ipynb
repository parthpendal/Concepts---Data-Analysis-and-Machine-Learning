```python
import cv2
import os
from google.colab.patches import cv2_imshow
img=cv2.imread("Boston_image1.jpg")
cv2_imshow(img)
```

```python
b,g,r=cv2.split(img)
cv2_imshow(b)
cv2_imshow(g)
cv2_imshow(r)
# for r image, red part will show as white
```

```python
#compare shapes
print(img.shape)
print(b.shape)
print(g.shape)
print(r.shape)
```

```python
#Merge these images
merged=cv2.merge([b,g,r])
cv2_imshow(merged)
```

```python
#Viewing r image as red (rather than seeing red part as while in grayscale image )
import numpy as np

blank=np.zeros(img.shape[:2],dtype='uint8')
blue=cv2.merge([b,blank,blank]) #Here b is blue channel, blank is green channel, blank is red channel
cv2_imshow(blue)

green=cv2.merge([blank,g,blank]) #Here g is green channel, blank is blue channel, blank is red channel
cv2_imshow(green)

red=cv2.merge([blank,blank,r]) 
cv2_imshow(red)
```
