```python
import cv2

#Accessing googledrive photos
from google.colab import drive
drive.mount('/content/gdrive')

import os
datapath="/content/gdrive/MyDrive/Face Mask detection"

#Getting directory for folders
folders=os.listdir(datapath)

#Getting path of first folder
path_of_first_folder=os.path.join(datapath,folders[1])

# Getting list of images name
Image_name_list=os.listdir(path_of_first_folder)

#Getting complete path of a image
path_of_first_image=os.path.join(path_of_first_folder,Image_name_list[222])

#Reading an image
first_read_image =cv2.imread(path_of_first_image)

from google.colab.patches import cv2_imshow
cv2_imshow(first_read_image)
```

```python
import numpy as np

#translation -- shifting image along x and y axis
def translate(img, x,y):
  transmat=np.float32([[1,0,x],[0,1,y]])
  dimensions=(img.shape[1],img.shape[0])
  return cv2.warpAffine(img,transmat,dimensions)

translated= translate(first_read_image, -30, 30)
cv2_imshow(translated)
```

```python
img=first_read_image.copy()

#Rotation
def rotate(img,angel,rotpoint=None):
  (height,width)= img.shape[:2]

  if rotpoint is None:
    rotpoint=(width/2,height/2)

  rotmat=cv2.getRotationMatrix2D(rotpoint,angel,1.0)
  dimensions=(width,height)

  return cv2.warpAffine(img,rotmat,dimensions)

rotated=rotate(img,45)
cv2_imshow(rotated)
```

```python
#resize a picture

resized=cv2.resize(img,(200,200),interpolation=cv2.INTER_CUBIC)
cv2_imshow(resized)
```

```python
#flip an image
flip=cv2.flip(img,-1)
#0 for flipping image vertically
#1 for flipping image horizontally
#-1 for fliiping image vertically as well ashorizontally 

cv2_imshow(flip)
```
