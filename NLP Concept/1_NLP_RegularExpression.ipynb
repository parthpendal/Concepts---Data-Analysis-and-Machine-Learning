```python
# prompt: find order number from chat4, chat5 and chat6

chat4 = 'My Order number is 1231231'
chat5 = 'my order# 12435323423'
chat6 = 'my order 21342342341'
pattern = 'order[^\d]*(\d*)| Order[^\d]*(\d*)'

matches = re.findall(pattern, chat4 + chat5 + chat6)

# Extract the order numbers from the matches
print(matches)
order_numbers = []
for match in matches:
    if match[0]:
        order_numbers.append(match[0])
    else:
        order_numbers.append(match[1])

# Print the order numbers
print(order_numbers)
```
